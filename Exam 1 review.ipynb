```python
%%bigquery --project data-viz-338818
SELECT *
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
LIMIT 50
```

flase alarm after 10pm to 5 am

```python

%%bigquery --project data-viz-338818
SELECT COUNT(*) as false_alarm_during_10pm_to_5am
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
where incident_group = "False Alarm" AND (hour_of_call >=22 OR hour_of_call <=4)
```

avarage

```python
%%bigquery --project data-viz-338818
SELECT incident_group, AVG(num_pumps_attending) as Average_pumps
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

max

```python
%%bigquery --project data-viz-338818
SELECT property_category, MAX(num_pumps_attending) as max_pumps
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY property_category
```
