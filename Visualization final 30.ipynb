```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

**BAR GRAPHS**

---



```python
ax = df.groupby('workout_type').calories.agg('mean').plot.barh()
ax.set(title = 'Average calories burn')
ax.set(xlabel = 'workout types')
plt.show()
```

**HISTOGRAM**

```python
ax = df.calories.hist()
ax.set(title = 'hostogram of calories')
plt.show()
```
