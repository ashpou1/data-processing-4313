##lets make histogram for goals

---



```python
df.Goals.plot.hist(title = 'Histogram of Goals',ylim=[0,450]).set_xlabel("Goals")
#https://github.com/ashpou1/data-processing-4313/blob/main/Histograms%2025.ipynb
```

##lets create a histogram of the Goals and Assists on the same axis.

---



```python
df[["Goals", "Assists"]].plot.hist(title = 'Histogram of Goals and Assists',ylim=[0,500]).set_xlabel("Goals and Assists")
```

##lets create a barchart for most goals scored by nationality

---



```python
ax = sns.barplot( x= df.groupby('nationality').Goals.agg('count').index, y = df.groupby('nationality').Goals.agg('count'))
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.annotate("Most Goals goes to Brazil",
            xy = [3,50],
            xytext = [4,60],
            arrowprops = dict(facecolor = 'gold', width = 3, headwidth = 12, headlength = 6))
ax.set_title('MOST GOAL SCORED BY NATIONALITY')
ax.set_xlabel('NATION')
ax.set_ylabel('GOALS')

plt.show()
#https://github.com/nurfnick/Data_Viz/blob/main/27_Proportions.ipynb
```

Wow, Brazil is the highest scored country in the given data of world cup.
