##lets make histogram for goals

---



```python
df.Goals.plot.hist(title = 'Histogram of Goals',ylim=[0,450]).set_xlabel("Goals")
#https://github.com/ashpou1/data-processing-4313/blob/main/Histograms%2025.ipynb
```

##lets create a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

---



```python
df[["Goals", "Assists"]].plot.hist(title = 'Histogram of Goals and Assists',ylim=[0,500]).set_xlabel("Goals and Assists")
```

##lets create a barchart for most goals scored by nationality

---



```python
ax = sns.barplot( x= df.groupby('nationality').Goals.agg('count').index, y = df.groupby('nationality').Goals.agg('count'))
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.annotate("Most Goals goes to Brazil",
            xy = [3,50],
            xytext = [4,60],
            arrowprops = dict(facecolor = 'gold', width = 3, headwidth = 12, headlength = 6))
ax.set_title('MOST GOAL SCORED BY NATIONALITY')
ax.set_xlabel('NATION')
ax.set_ylabel('GOALS')

plt.show()
#https://github.com/nurfnick/Data_Viz/blob/main/27_Proportions.ipynb
```

Wow, Brazil is the highest scored country in the given data of world cup.

##lets create ‘Goals’ and ‘Assists’ as scatter plot
---



```python
ax = sns.scatterplot(data = df, 
                x = 'Goals', 
                y = 'Assists',
                hue = 'club',s=100)

sns.move_legend(ax, "upper right", bbox_to_anchor=(-.2, 1))
```

##lets create stacked barplot of goals and assists of French National Team.

---



```python
df1 = df[df["nationality"] == 'France' ]
df1.head()
```

```python
ax = df1.groupby(['Goals', 'Assists']).size().unstack().plot(kind='bar', stacked=True)
ax.set(title= 'Stacked Goals and Assist of french Team')
ax.set_xlabel('Goals')
ax.set_ylabel('Assists')
plt.show()
#https://stackoverflow.com/questions/23415500/pandas-plotting-a-stacked-bar-chart
```

I think this shows the relation how many goals are depends in assists. As i see throught color and bars like french get 22 goals after 11 assists

##lets create a correlation heatmap of the entire dataset

---



```python
sns.heatmap(df.corr(), annot=True, linewidths=0.25,vmin = -2)
plt.title('Correlation heatmap of world cup')
plt.show()
#https://console.cloud.google.com/bigquery?project=data-viz-338818
```
